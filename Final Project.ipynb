```python
cFIP_dict = {}

for index, row in fan_graphs_df.iterrows():
  cFIP_dict[int(row['Season'])] = row['cFIP']

wHBP_dict = {}

for index, row in fan_graphs_df.iterrows():
  wHBP_dict[int(row['Season'])] = row['wHBP']
```

```python
mlb_stats['1B'] = mlb_stats['H'] - (mlb_stats['2B'] + mlb_stats['3B'] + mlb_stats['HR'])

mlb_stats['BA'] = mlb_stats['H'] / mlb_stats['AB']
mlb_stats['OBP'] = (mlb_stats['H'] + mlb_stats['BB'] + mlb_stats['HBP']) / (mlb_stats['AB'] + mlb_stats['BB'] + mlb_stats['SF'] + mlb_stats['HBP'])
mlb_stats['SLG'] = (mlb_stats['1B'] + 2 * mlb_stats['2B'] + 3 * mlb_stats['3B'] + 4 * mlb_stats['HR']) / mlb_stats['AB']
mlb_stats['OPS'] = mlb_stats['OBP'] + mlb_stats['SLG']
mlb_stats['WHIP'] = (mlb_stats['BBA'] + mlb_stats['HA']) / (mlb_stats['IPouts'] / 3)
mlb_stats['adjFIP'] = (((13 * mlb_stats['HRA']) + (3 * mlb_stats['BBA'] * mlb_stats['yearID'].map(wHBP_dict)) - (2 * mlb_stats['SOA'])) /
                       ((mlb_stats['IPouts'] / 27) + mlb_stats['yearID'].map(cFIP_dict)))

#Traditionally, FIP includes the total number of batters hit by picthes, but this dataset doesn't have that metric, hence the "adjFIP.
#Instead of adding HBP per year, we multiplied 'BBA' by wHBP per year from the FanGraphs dataset I found above (where we also got the FIP constant)
#to make our FIP values align a little more closely with what FIP values would actually look like


mlb_stats[['BA', 'OBP', 'SLG', 'OPS', 'WHIP', 'adjFIP']].tail(10)
```

```python
numeric_mlb_stats = mlb_stats.select_dtypes(include=[np.number])
correlation_matrix = numeric_mlb_stats.corr()

collinear_pairs = []
for row in correlation_matrix.index:
    for column in correlation_matrix.columns:
        if row < column and abs(correlation_matrix.loc[row, column]) >= 0.7:
            collinear_pairs.append((row, column, correlation_matrix.loc[row, column]))

collinear_df = pd.DataFrame(collinear_pairs, columns=['Feature 1', 'Feature 2', 'Correlation'])

collinear_df
```

# Multiple Regression Model

```python
# predictors with high correlation to wins while taking account collinearity
predictors = ['R', 'IPouts', 'BB', 'SHO', 'SF', '2B', '1B', 'SV', 'FP', 'attendance', 'OBP']

selected_data = numeric_mlb_stats[predictors]
selected_correlation_matrix = pd.DataFrame(selected_data.corr())
selected_correlation_matrix
```

```python
# Model for the entire dataset

from sklearn.metrics import r2_score

mlb_stats_clean = mlb_stats.dropna(subset=predictors + ['W'])

x = mlb_stats_clean[predictors]
y = mlb_stats_clean['W']

model = LinearRegression()
model.fit(x, y)
intercept = model.intercept_
coefficients = model.coef_

predictions = model.predict(x)
r_squared = r2_score(y, predictions)

n = len(y)
k = len(coefficients)
adjusted_r2 = 1 - ((1 - r_squared) * (n - 1) / (n - k - 1))

equation_parts = [f'{coef} * {feature}' for coef, feature in zip(coefficients, x.columns)]
regression_equation = f'yÃÇ = {intercept} + ' + ' + '.join(equation_parts)

print(regression_equation)
print(f'Adjusted R^2 = {adjusted_r2}')
```

```python
# Predicting Yankees 2023 Stats as a sanity check for the model, we used approximations for some datapoints just to get an idea
data = {
    'R': [769],
    'IPouts': [4319],
    'BB': [508],
    'SHO': [3],
    'SF': [50],
    '2B': [280],
    '1B': [900],
    'SV': [44],
    'FP': [0.987],
    'attendance': [3269016],
    'OBP': [0.317]
}

# Convert the data to a DataFrame
team_data = pd.DataFrame(data)

# Create and fit the regression model (assuming 'model' is already defined and fitted with your dataset)
predicted_wins = model.predict(team_data)

# Output the predicted number of wins
print(f'Predicted Wins for the New York Yankees in 2023: {predicted_wins[0]:.0f}')
print(f"Actual Wins for the New York Yankees in 2023: {mlb_stats.loc[(mlb_stats.yearID == 2023) & (mlb_stats.name == 'New York Yankees')]['W'].values[0]}")
```

```python
# Testing models on the most recent 5 years

def train_and_evaluate(X_train, y_train, X_test, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    return r2_score(y_test, predictions)

most_recent_year = mlb_stats['yearID'].max()
test_data = mlb_stats[(mlb_stats['yearID'] > most_recent_year - 5) & (mlb_stats['yearID'] <= most_recent_year)]
X_test = test_data[predictors]
y_test = test_data['W']

mlb_stats_train = mlb_stats[mlb_stats['yearID'] <= most_recent_year - 5]

mlb_stats_clean = mlb_stats_train.dropna(subset=predictors + ['W'])
X_train_all = mlb_stats_clean[predictors]
y_train_all = mlb_stats_clean['W']

# World Series Era (1903-present)
mlb_stats_era = mlb_stats_train[mlb_stats_train['yearID'] >= 1903]
mlb_stats_clean_era = mlb_stats_era.dropna(subset=predictors + ['W'])
X_train_era = mlb_stats_clean_era[predictors]
y_train_era = mlb_stats_clean_era['W']

# Integrated Era (1947-present)
mlb_stats_integrated = mlb_stats_train[mlb_stats_train['yearID'] >= 1947]
mlb_stats_clean_integrated = mlb_stats_integrated.dropna(subset=predictors + ['W'])
X_train_integrated = mlb_stats_clean_integrated[predictors]
y_train_integrated = mlb_stats_clean_integrated['W']

# Instant Replay Era (2008-present)
mlb_stats_replay = mlb_stats_train[mlb_stats_train['yearID'] >= 2008]
mlb_stats_clean_replay = mlb_stats_replay.dropna(subset=predictors + ['W'])
X_train_replay = mlb_stats_clean_replay[predictors]
y_train_replay = mlb_stats_clean_replay['W']

r2_all = train_and_evaluate(X_train_all, y_train_all, X_test, y_test)
r2_era = train_and_evaluate(X_train_era, y_train_era, X_test, y_test)
r2_integrated = train_and_evaluate(X_train_integrated, y_train_integrated, X_test, y_test)
r2_replay = train_and_evaluate(X_train_replay, y_train_replay, X_test, y_test)

print(f"Entire Dataset R^2 on Test Data: {r2_all}")
print(f"World Series Era R^2 on Test Data: {r2_era}")
print(f"Integrated Era R^2 on Test Data: {r2_integrated}")
print(f"Instant Replay Era R^2 on Test Data: {r2_replay}")
```

```python
r2_scores = [r2_all, r2_era, r2_integrated, r2_replay]
labels = ['Entire Dataset (1871)', 'World Series Era (1903)', 'Integrated Era (1947)', 'Instant Replay Era (2008)']

plt.figure(figsize=(10, 6))
plt.plot(labels, r2_scores, marker='o', color='b', linestyle='-')
plt.xlabel('Era')
plt.ylabel('R^2 Score')
plt.title('R^2 Scores of Models on Test Data')
plt.ylim(-5, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.model_selection import KFold

x = mlb_stats_clean[predictors_final]
y = mlb_stats_clean['W']

def cross_val_score_predictors(x, y, predictors, n_splits=5, shuffle=True, random_state=None):
    kf = KFold(n_splits=n_splits, shuffle=shuffle, random_state=random_state)
    scores = []
    for train_index, test_index in kf.split(x):
        x_train, x_test = x.iloc[train_index][predictors], x.iloc[test_index][predictors]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model = LinearRegression()
        model.fit(x_train, y_train)
        predictions = model.predict(x_test)
        score = r2_score(y_test, predictions)
        scores.append(score)
    return np.mean(scores)

# We can change this number of iterations to run the feature selection process
num_iterations = 30
best_overall_score = -np.inf
best_overall_predictors = []

for i in range(num_iterations):
    best_predictors = []
    remaining_predictors = predictors_final[:]
    best_score = -np.inf

    while remaining_predictors:
        improvement = False
        for predictor in remaining_predictors:
            current_predictors = best_predictors + [predictor]
            score = cross_val_score_predictors(x, y, current_predictors, random_state=i)
            if score > best_score:
                best_score = score
                best_predictor = predictor
                improvement = True
        if improvement:
            best_predictors.append(best_predictor)
            remaining_predictors.remove(best_predictor)
        else:
            break

    print(f"Iteration {i+1}: Best predictors = {best_predictors}, Best R^2 Score = {best_score}")

    if best_score > best_overall_score:
        best_overall_score = best_score
        best_overall_predictors = best_predictors

print("Best overall set of predictors:", best_overall_predictors)
print("Best overall R^2 Score:", best_overall_score)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier


predictors = [('OPS', 'BA'), ('OPS', 'H'), ('BA', 'H')]

y = mlb_stats_finalkNN['DivWin_Y']

for pair in predictors:
    X = mlb_stats_finalkNN[list(pair)]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

    k_values = range(1, 13)

    cv_scores_list = []
    test_accuracies = []

    for k in k_values:
        # Initialize KNeighborsClassifier with current k
        knn = KNeighborsClassifier(n_neighbors=k)

        # Perform cross-validation
        cv_scores = cross_val_score(knn, X_train_smote, y_train_smote, cv=5)
        mean_cv_score = np.mean(cv_scores)
        cv_scores_list.append(mean_cv_score)

        knn.fit(X_train_smote, y_train_smote)

        test_accuracy = knn.score(X_test, y_test)
        test_accuracies.append(test_accuracy)

    plt.figure(figsize=(10, 5))
    plt.plot(k_values, cv_scores_list, label='Mean CV Score', marker='o')
    plt.plot(k_values, test_accuracies, label='Test Accuracy', marker='x')
    plt.title(f'SMOTE {pair[0]} vs {pair[1]}')
    plt.xlabel('Number of Neighbors: k')
    plt.ylabel('Accuracy')
    plt.xticks(k_values)
    plt.legend()
    plt.grid(True)
    plt.show()
```

I think the most interesting insights that we gained from constructing these 24 models is this: there was a clear pattern present that supported the usage of more advanced, SABrmetric stats (OPS and adjFIP) in predicting DV outcomes (regular season success) and the usage of less advanced, results-based stats in predicting WS outcomes (postseason success). This is a super interesting phenomenon that makes sense, intuitively. These advanced stats are indicative of underlying player and team performance capabilities that may not always directly translate into immediate wins but do contribute significantly to a team‚Äôs overall ability and consistency across a regular season. On the other hand, less advanced, results-based stats like earned run average (ERA) and batting average (BA) tend to capture the outcomes rather than the processes, which might be more influential in the high-pressure, high-variance context of postseason games.

This distinction between the types of statistics that are more predictive for regular season versus postseason success is insightful because it highlights how different aspects of baseball are valued and can be predictive in different contexts. Regular season games, which are more numerous, might allow the subtleties of SABRmetric stats like OPS (On-base Plus Slugging) and adjFIP (Adjusted Fielding Independent Pitching) to play out and significantly impact a team‚Äôs standing over time. These stats are designed to remove some of the randomness and luck involved in traditional stats, providing a clearer picture of a player's or team's true skill level.

In contrast, the postseason's shorter series format and higher stakes mean that immediate results‚Äîsuch as runs batted in or earned runs allowed‚Äîcan have a disproportionately large impact. In such scenarios, the ability to capitalize on or respond to specific situations (which results-based stats capture) can be more directly correlated with winning games and advancing in the playoffs.

Overall, these insights suggest that teams might benefit from tailoring their analytic approaches depending on the part of the season they are strategizing for. Emphasizing advanced metrics during the player acquisition and development phases could build a strong foundation for regular season success, while a focus on maximizing results-based performance might be more critical when preparing for and competing in the postseason. This nuanced understanding of statistical relevance in different baseball contexts can help teams allocate resources more effectively and develop strategies that are aligned with the unique demands of regular season and postseason play.
